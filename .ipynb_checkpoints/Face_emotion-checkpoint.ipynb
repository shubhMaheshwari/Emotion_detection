# Detect Emotions 

```python
%matplotlib notebook

# For getting landmarks
import face_detect
import face_recognize
from imutils import face_utils,resize

# For displaying
import cv2
from matplotlib import pyplot as plt

# For loading images
import os
from PIL import Image

# Proccessing
import pandas
import numpy as np
```

```python
# Return landmarks as numpy array
def np_landmark(frame, verbose=False):
    #         Detect the face
        face_locations = face_detect.detect_face(frame)

        #         Get the landmarks
        raw_landmarks = face_recognize._raw_face_landmarks(frame, face_locations, model="large")
        landmarks_np = face_utils.shape_to_np(raw_landmarks)
        
        if verbose == True:
            for mark in landmarks_np:
                x,y = mark
                cv2.circle(frame,(x, y), 1, (0, 0, 255), -1)

            (face_start_x, face_start_y, face_end_x, face_end_y) = face_locations
            cv2.rectangle(frame, (face_start_x, face_start_y ), (face_end_x, face_end_y ),(255,0,0),2)
        
        return landmarks_np/frame.shape[:2]
```

# Load images and their predictions 

```python
df = pandas.read_csv('datasets/facial_expressions/data/legend.csv')
print('Emotion', df['emotion'].str.upper().unique())
```

```python
IMG_EXTENSIONS = [
    '.jpg', '.JPG', '.jpeg', '.JPEG',
    '.png', '.PNG', '.ppm', '.PPM', '.bmp', '.BMP',
]


def is_image_file(filename):
    return any(filename.endswith(extension) for extension in IMG_EXTENSIONS)


def default_loader(path):
    return np.array(Image.open(path).convert('RGB'))

def make_dataset(dir):
    images = []
    emotions = []
    assert os.path.isdir(dir), '%s is not a valid directory' % dir

#     Now use Y
    for i, (id, fname, emotion) in enumerate(df.values):
        if is_image_file(fname):
            path = os.path.join(dir, fname)
            RGB = default_loader(path)
            try:
                landmarks = np_landmark(RGB)
                images.append(landmarks)
                emotions.append(emotion)
            except:
                pass
        if i % 1000 == 0:
            print("Completed:", i)
    return np.array(images), np.array(emotions)

```

# Get X and Y for prediction

```python
try:
    data = pandas.read_csv('landmarks_emotion.csv')
    data['0.1'] = data['0.1'].str.upper()
except FileNotFoundError as f:
    X,y = make_dataset('./datasets/facial_expressions/images/')
    print("No of images",len(X))
    dat1 = pandas.DataFrame(X.reshape((X.shape[0],68*2)))
    dat2 = pandas.DataFrame(y)
    dat3 = pandas.concat([dat1, dat2], axis=1)
    dat3.to_csv("landmarks_emotion.csv")
    data = dat3
```

# Train our MLP

```python
from sklearn.neural_network import MLPClassifier  
from sklearn.model_selection import train_test_split
from sklearn import preprocessing  
le = preprocessing.LabelEncoder()

from sklearn.metrics import classification_report, confusion_matrix  
```

```python
X = data.iloc[:,1:-1]
y = data.select_dtypes(include=[object])
```

```python
y_pred = y.apply(le.fit_transform)  
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_pred, test_size = 0.20)  
```

```python
mapping = ['ANGER','CONTEMPT', 'DISGUST', 'FEAR', 'Happy', 'Neutral', 'SUPRISE']
```

# Our Network

```python
from keras.models import Sequential
from keras.layers import Dense
from  keras.optimizers import Adam 
from keras import losses
from matplotlib import pyplot
```

```python
model = Sequential()
model.add(Dense(128,input_dim=136,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(8,activation='sigmoid'))
```

```python
adam =Adam(lr=0.0001, beta_1=0.9, beta_2=0.999)
model.compile(loss=losses.categorical_crossentropy, optimizer=adam, metrics=['accuracy'])
```

```python
# from labels get the hot vector
yy_train = np.zeros((y_train.shape[0],8))
yy_train[np.arange(y_train.shape[0]),y_train.values.ravel()] = 1
history = model.fit(X_train.values, yy_train, epochs=500, batch_size=3000, verbose=4)
```

```python
pyplot.plot(history.history['acc'])
pyplot.show()
```

```python
model.predict(X_train.values)
```

```python
y_test_pred = np.argmax(model.predict(X_train.values),axis=1)
```

```python
# Run webcam, for later testing
def webcam_run():
    video_capture = cv2.VideoCapture(0)

    while(True):
        ret, frame = video_capture.read()
        try:
            landmarks = np_landmark(frame,True)
            ind = np.argmax(model.predict(landmarks.reshape((1,136))))
            print(mapping[ind])
            cv2.imshow('frame',frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        except Exception as e:
            print(e)
            pass
    video_capture.release()
    cv2.destroyAllWindows()
webcam_run()
```
